```python
import pandas as pd

header = ["f1", "f2", "c"]
data = pd.read_csv("HW3_Q3_1.csv", names=header)
```

```python
print(data.c.unique())
```

```python
from sklearn.neighbors.nearest_centroid import NearestCentroid

clf = NearestCentroid()
clf.fit(data[["f1", "f2"]], data["c"])
centroids = clf.centroids_
print(centroids)
```

```python
import matplotlib.pyplot as plt

plt.scatter(data.f1, data.f2, c=data.c)
plt.scatter(centroids[:,0],centroids[:,1], c="black", marker="^")
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans()
kmeans.set_params(n_clusters=7)
y_km = kmeans.fit(data[["f1", "f2"]])
```

```python
plt.scatter(data.f1, data.f2, c=data.c)
plt.scatter(centroids[:,0],centroids[:,1], c="black", marker="^")
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], marker="+", s=1000)
```
